HTML TEXT SPLITTER


```python
from langchain_text_splitters import HTMLHeaderTextSplitter
html_string = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Generative AI Demo</title>
    <style>
        body { 
            font-family: Arial, sans-serif; 
            background-color: #f5f5f5; 
            margin: 40px; 
        }
        h1 { 
            color: #4A90E2; 
        }
        p { 
            color: #333; 
            line-height: 1.6; 
        }
        .highlight {
            background-color: #ffef96;
            padding: 4px;
            border-radius: 4px;
        }
    </style>
</head>
<body>
    <h1>Welcome to Generative AI</h1>
     <h2>Welcome to Generative AI</h2>
      <h3>Welcome to Generative AI</h3>
    <p>This is a <span class="highlight">demo page</span> for testing HTML loading with LangChain.</p>
    <p>Generative AI is transforming the world by enabling machines to create <strong>text</strong>, <em>images</em> and even <u>code</u>!</p>
</body>
</html>
"""

headers_to_split_on = [
    ("h1" , "Header 1"),
    ("h3" , "Header 3")
]


html_splitter = HTMLHeaderTextSplitter(headers_to_split_on)

final_html = html_splitter.split_text(html_string)
print(final_html)
```
